```python
# Open Library
```

```python
import pandas as pd
import requests 
from pandas import json_normalize
```

```python
res_library = requests.get('https://openlibrary.org/api/books?bibkeys=ISBN%3A9781408113479%2COCLC%3A420517&format=json&jscmd=viewapi')
```

```python
open_library_json = res_library.json()
```

```python
open_library_data = pd.DataFrame(open_library_json)
open_library_data.T.reset_index(drop=True)
```
